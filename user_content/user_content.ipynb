```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('user.csv')
```

```python
df.head(5)
```

```python
del df['user_id']
del df['name']

st = df['review_count'].std()
me = df['review_count'].mean()
df['review_count'] = (df['review_count'] - me)/st

df['yelping_since'] = df['yelping_since'].astype(str).str[0:4]
del df['friends']

st = df['useful'].std()
me = df['useful'].mean()
df['useful'] = (df['useful'] - me)/st

st = df['funny'].std()
me = df['funny'].mean()
df['funny'] = (df['funny'] - me)/st

st = df['cool'].std()
me = df['cool'].mean()
df['cool'] = (df['cool'] - me)/st

st = df['fans'].std()
me = df['fans'].mean()
df['fans'] = (df['fans'] - me)/st

del df['elite']

stri = 'average_stars'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st

stri = 'compliment_hot'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st

stri = 'compliment_more'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st

stri = 'compliment_profile'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st

stri = 'compliment_cute'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st

stri = 'compliment_list'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st

stri = 'compliment_note'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st

stri = 'compliment_plain'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st

stri = 'compliment_cool'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st

stri = 'compliment_funny'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st

stri = 'compliment_writer'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st

stri = 'compliment_photos'
st = df[stri].std()
me = df[stri].mean()
df[stri] = (df[stri] - me)/st
```

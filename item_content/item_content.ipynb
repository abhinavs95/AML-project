```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('business.csv')
```

```python
df.head(5)
```

```python
del df['hours.Monday']
del df['hours.Tuesday']
del df['hours.Wednesday']
del df['hours.Thursday']
del df['hours.Friday']
del df['hours.Saturday']
del df['hours.Sunday']
st = df['review_count'].std()
me = df['review_count'].mean()
df['review_count'] = (df['review_count'] - me)/st
st = df['latitude'].std()
me = df['latitude'].mean()
df['latitude'] = (df['latitude'] - me)/st
st = df['longitude'].std()
me = df['longitude'].mean()
df['longitude'] = (df['longitude'] - me)/st
del df['name']
del df['address']
del df['postal_code']
del df['categories']
del df['business_id']
df['is_open'] = df['is_open'].astype('str')
df['stars'] = df['stars'].astype('str')
df['attributes.RestaurantsPriceRange2'] = df['attributes.RestaurantsPriceRange2'].astype('str')
```

```python
df.head(5)
```

```python
res = pd.get_dummies(df, dummy_na=False)
```

```python
res.head(5)
```

```python
res.to_csv('output.csv')
```
